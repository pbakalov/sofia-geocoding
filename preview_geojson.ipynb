## load geodata

```python
import geopandas as gpd
```

```python
import geojson 
```

```python
data_dir = './'

# uncomment one 
file = 'sofia_regions.geojson' # admin regions, from OSM, see https://community.openstreetmap.org/t/topic/81562 
# file = 'sofia_city_units.json' # neighborhoods (sort of), from https://ge.sofiaplan.bg/
# file = 'kvartali_26_sofpr_20190101.geojson' # neighborhoods, from https://sofiaplan.bg/api/
```

```python
with open(f'{data_dir}/{file}', "r", encoding="utf-8") as f:
    geodata = geojson.load(f)
```

```python
geodata['features'][0]['properties'] 
```

## quick preview with folium 

```python
import folium 
```

```python
center = [42.75, 23.6]
```

```python
m = folium.Map(location=center, zoom_start=10)
```

```python
# Add the GeoJSON data to the map with tooltips
folium.GeoJson(
    geodata,
    name='geojson',
    tooltip=folium.GeoJsonTooltip(
        fields=['name'], 
#         fields=['kvname'], 
        aliases=['Location Name:']
    ),
).add_to(m)
```
